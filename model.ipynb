```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies=pd.read_csv('indian movies.csv')
movies.info()
```

```python
movies.shape
```

```python
movies.replace('-', pd.NA, inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.isnull().sum()

movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.head()
```

```python
movies = movies[['Movie Name', 'Genre', 'Language', 'Year']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies['tags'] = (
    movies['Genre'] + " " +
    movies['Language'] + " " +
    movies['Year'].astype(str)
)
```

```python
movies['tags']=movies['tags'].str.lower()
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
print(vectors.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
print(similarity.shape)
```

```python
def recommend(movie):

    if movie not in movies['Movie Name'].values:
        print("Movie not found")
        return

    movie_index = movies[movies['Movie Name'] == movie].index[0]

    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    for i in movies_list:

        print(
            movies.iloc[i[0]]['Movie Name'],
            "-> Similarity Score:",
            round(i[1], 2)
        )
```

```python
recommend('3 Idiots')
```
